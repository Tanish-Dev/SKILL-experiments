# **Experiment 3**

```python
#Label Enccoding

import pandas as pd
from sklearn.preprocessing import LabelEncoder

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df5 = pd.read_csv(url,header=None, names=column_names)

print("Original Dataset")
print(df5.head())


label_encoder = LabelEncoder()
df5['Species_LabelEconded'] = label_encoder.fit_transform(df5['species'])

print("\nDataset after Label Encoding")
df5.head()
```

```python
# One Hot Encoding

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df5 = pd.read_csv(url,header=None, names=column_names)

print("Original Dataset")
print(df5.head())

encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(df5[['species']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['species']))
df_one_hot_encoded = pd.concat([df5.drop('species', axis=1), encoded_df], axis=1)

print("\nDataset after One Hot Encoding")
print(encoded_df.head())
```

```python
df2 = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red']
})

encoded_df = pd.get_dummies(df2).sum()
print(encoded_df)
```

```python
df2 = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red']
})
encoded_df = pd.get_dummies(df2, columns=['Color'],dtype=int)
print(encoded_df)
```

# **Experiment 4**

```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler, StandardScaler


iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

mini_max_scaler = MinMaxScaler()
df_mini_max_scaled = mini_max_scaler.fit_transform(df)

standard_scaler = StandardScaler()
df_standard_scaled = standard_scaler.fit_transform(df)

df_mini_max_scaled = pd.DataFrame(df_mini_max_scaled, columns=df.columns)
df_standard_scaled = pd.DataFrame(df_standard_scaled, columns=df.columns)

print("Original Dataset:")
print(df.head())
print("\nDataset after Min-Max Scaling:")
print(df_mini_max_scaled.head())
print("\nDataset after Standard Scaling:")
print(df_standard_scaled.head())
```
